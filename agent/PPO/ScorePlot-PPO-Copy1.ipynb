```python
import matplotlib.pyplot as plt
import numpy as np
import torch
%matplotlib inline
```

```python
scores = torch.load('plot/5000_scores.pth')
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 6))
ax.grid(True)
ax.plot(scores);
```

```python
window_size = 30
running_mean_scores = [np.mean(scores[max(i-window_size+1, 0):i+1]) for i in range(len(scores))]
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 6))
ax.grid(True)
ax.plot(running_mean_scores, color='orange');
```
