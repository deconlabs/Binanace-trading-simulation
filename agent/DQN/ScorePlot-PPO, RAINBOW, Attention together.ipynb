```python
import matplotlib.pyplot as plt
import numpy as np
import torch
%matplotlib inline
```

```python
ppo_scores = torch.load('PPO/saves/Original/1/3000_scores.pth')
rainbow_scores = torch.load('DQN/saves/Original/1/scores.pth')
attention_scores = torch.load('Attention/saves/1/scores.pth')
```

```python
def draw_scores(win_size, scores,name):
    window_size = win_size

    running_mean_scores = [np.mean(scores[max(i-window_size+1, 0):i+1]) for i in range(len(scores))]

    score_mean_std   = (np.mean(scores), np.std(scores) * 1.5)
#    running_mean_std = (np.mean(running_mean_scores), np.std(running_mean_scores) * 1.5)
    
#     fig, ax = plt.subplot(1,1)
    plt.plot(running_mean_scores, label = name)
#     ax[0].set_ylim(running_mean_std[0] - running_mean_std[1], running_mean_std[0] + running_mean_std[1])

```

```python
plt.plot(ppo_scores,label = "PPO")    
plt.plot(rainbow_scores, label = "RAINBOW")
plt.plot(attention_scores, label = "ATTENTION")
plt.grid()
plt.legend()
```

```python
for scores,name in [(ppo_scores,'PPO'), (rainbow_scores,"RAINBOW"), (attention_scores,"ATTENTION")]:
    draw_scores(100,scores, name)
plt.grid()
plt.legend()
plt.savefig("TradingAgentPerformance.png",dpi = 300)
```
